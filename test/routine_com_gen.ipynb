```python
filts = np.arange(400, 900, 25)
units = np.arange(1, 21)
datelist = []
files = []

# /ldata/data1/obsdata/7DT06/2024-09-20_gain2750
for unit in units:
    print("="*23+'\n', f"7DT{unit:02}", '\n'+"="*23)
    for tile in gw190814_tiles:
        src = f'/lyman/data1/obsdata/7DT{unit:02}/2024-??-??_gain2750/*T{tile:05}*.fits'
        globbed = glob(src)
        if len(globbed) == 0:
            # print("No Files")
            continue
        count = 0
        for fname in globbed:
            hdr = fits.getheader(fname)
            if hdr['NOTE'] == 'GW190814':
                count += 1
                datelist.append(hdr['DATE-LOC'])
                files.append(fname)
        # print(count)
```
