```python
import os
import subprocess
import glob
import numpy as np
from astropy.table import Table
```

```python
path_data = "/lyman/data1/processed_1x1_gain2750"
path_ref = "/lyman/data1/factory/ref_frame"
```

```python
tiles = glob.glob(f"{path_data}/T?????")
print(f"{len(tiles)} tiles found")
```

```python
tile = tiles[1]
print(tile)
```

```python
imlist = glob.glob(f"{tile}/7DT??/?/c*m.fits")
print(f"{len(imlist)} images found")
```

```python
f = open(f"{path_ref}/no_phot_cats.txt", "w")

for tt, tile in enumerate(tiles):
	imlist = glob.glob(f"{tile}/7DT??/?/c*m.fits")
	print(f"[{tt+1:0>4}/{len(tiles):0>4}] {len(imlist)} images found in {tile}")
	if len(imlist) > 0:
		for inim in imlist:
			#
			photcat = f"{os.path.dirname(inim)}/phot/{os.path.basename(inim).replace('.fits', '.phot.cat')}"
			#	Info
			filte = os.path.basename(inim).split("_")[-2]
			#	To the output path
			path_in = os.path.dirname(inim)
			path_out = f"{path_ref}/{filte}"
			outim = inim.replace(path_in, path_out)
			outcat = photcat.replace(f"{path_in}/phot", path_out)
			#	Make the directory
			os.makedirs(path_out, exist_ok=True)

			
			#	Command
			cpcom_image = ['cp', inim, outim]
			subprocess.run(cpcom_image, check=True)

			if os.path.exists(photcat):
				f.write(f"{photcat}\n")
				# continue
				cpcom_cat = ['cp', photcat, outcat]
				subprocess.run(cpcom_cat, check=True)
	else:
		pass

f.close()
```
